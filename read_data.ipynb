```python
import pandas as pd
import seaborn as sns
import pylab as pl
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_dir = "../Lucas_Cecile/"
```

```python
data = pd.read_excel(data_dir+"PST_databrutes.xlsx")
```

```python
data
```

```python
columns_ref = np.array(['A','B','C','D','E'])
rows_ref = np.arange(1,6)
```

```python
processed_data = pd.DataFrame()
for grp in data.groupby(['Subject','Session','Phase','Species']):
    print(grp[0])
#     print(grp[1])
    temp = pd.DataFrame()
    sub = grp[1].loc[(grp[1]['Modifier #1']=='Success')|(grp[1]['Modifier #1']=='Revisit')]
    coords = [[ rows_ref[np.where(rows_ref==int(x[1]))[0]]-1, np.where(columns_ref==x[0])[0] ]  for x in sub['Behavior'].values]
    rows = np.array(coords)[:,0].flatten()
    cols = np.array(coords)[:,1].flatten()
    
    temp['Behavior'] = sub['Behavior'].values
    temp['Phase_Session'] = sub['Phase_session'].values
    temp['Rows'] = rows
    temp['Cols'] = cols
    temp['Reward'] = [1 if x == 'Success' else 0 for x in sub['Modifier #1'].values]
    temp['Time'] = sub['Time'].values
    #temp['Species'] = sub['Species'].values
    temp['PST_condition'] = sub['PST condition'].values
    temp['Subject'] = grp[0][0]
    temp['Session'] = grp[0][1]
    temp['Phase'] = grp[0][2]
    temp['Species'] = grp[0][3]
    processed_data = pd.concat([processed_data,temp])
```

```python
processed_data.to_csv(data_dir+"extracted_trajs.csv")
```
